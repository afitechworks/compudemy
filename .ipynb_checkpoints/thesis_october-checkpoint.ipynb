```python
import numpy as np
import pandas as pd
from geopy.distance import geodesic
import random
import math

# ---------- PARAMETERS ----------
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

# fraction of rows to mark malicious overall
MALICIOUS_FRACTION = 0.05   # 5% overall (tune as needed)

# specific fractions (relative or absolute)
GPS_SPOOF_FRAC = 0.25
REPLAY_FRAC = 0.15
FABRICATION_FRAC = 0.15
SYBIL_FRAC = 0.10
TIMESTAMP_SHIFT_FRAC = 0.10
SPAT_SPOOF_FRAC = 0.10
DOS_FRAC = 0.10
SUBTLE_MOD_FRAC = 0.05

# realistic thresholds
MAX_REALISTIC_SPEED_KMH = 180.0
SPOOF_DISTANCE_MIN_M = 200
SPOOF_DISTANCE_MAX_M = 2000
FABRICATED_SPEED_MIN_KMH = 220
FABRICATED_SPEED_MAX_KMH = 400

# ---------- helpers ----------
def meters_between(lat1, lon1, lat2, lon2):
    try:
        return geodesic((lat1, lon1), (lat2, lon2)).meters
    except:
        return np.nan

def kmh_from_m_s(m_s):
    return m_s * 3.6

def move_point(lat, lon, bearing_deg, distance_m):
    # returns new lat, lon given bearing (deg) and distance (m)
    R = 6371000.0
    br = math.radians(bearing_deg)
    lat1 = math.radians(lat)
    lon1 = math.radians(lon)
    dR = distance_m / R
    lat2 = math.asin(math.sin(lat1)*math.cos(dR) + math.cos(lat1)*math.sin(dR)*math.cos(br))
    lon2 = lon1 + math.atan2(math.sin(br)*math.sin(dR)*math.cos(lat1), math.cos(dR)-math.sin(lat1)*math.sin(lat2))
    return math.degrees(lat2), (math.degrees(lon2)+180)%360 - 180

# ---------- prepare df ----------
df = df_clean.copy()  # use your actual merged dataframe here
n = len(df)
print("Rows:", n)

# initialize attack columns
df['is_malicious'] = 0
df['attack_type'] = np.nan

attack_flags = ['gps_spoof_flag','replay_flag','fabrication_flag','sybil_flag',
                'timestamp_shift_flag','spat_spoof_flag','dos_flag','subtle_mod_flag']
for c in attack_flags:
    df[c] = 0

# ---------- compute baseline features ----------
df['metadata_generatedAt'] = pd.to_datetime(df['metadata_generatedAt'], errors='coerce')
df = df.sort_values(['coreData_id','metadata_generatedAt']).reset_index(drop=True)
df['prev_lat'] = df.groupby('coreData_id')['coreData_position_lat'].shift(1)
df['prev_long'] = df.groupby('coreData_id')['coreData_position_long'].shift(1)
df['prev_time'] = df.groupby('coreData_id')['metadata_generatedAt'].shift(1)
df['prev_speed'] = df.groupby('coreData_id')['coreData_speed'].shift(1)

def compute_pos_jump(row):
    if pd.isna(row['prev_lat']) or pd.isna(row['prev_long']) or pd.isna(row['metadata_generatedAt']) or pd.isna(row['prev_time']):
        return np.nan
    dt = (row['metadata_generatedAt'] - row['prev_time']).total_seconds()
    if dt <= 0:
        return np.nan
    return meters_between(row['prev_lat'], row['prev_long'], row['coreData_position_lat'], row['coreData_position_long'])

df['pos_jump_m'] = df.apply(compute_pos_jump, axis=1)
df['time_delta_s'] = (df['metadata_generatedAt'] - df['prev_time']).dt.total_seconds()
df['computed_speed_kmh'] = (df['pos_jump_m'] / df['time_delta_s']).replace([np.inf, -np.inf], np.nan) * 3.6
df['speed_inconsistent_kmh'] = np.abs(df['coreData_speed'] - df['computed_speed_kmh'])
df['duplicate_msg'] = df.duplicated(subset=['coreData_id','coreData_msgCnt','coreData_secMark'], keep=False).astype(int)
df['msg_epoch_min'] = df['metadata_generatedAt'].dt.floor('T')
msg_rate = df.groupby(['coreData_id','msg_epoch_min']).size().reset_index(name='count_per_min')
df = df.merge(msg_rate, on=['coreData_id','msg_epoch_min'], how='left')
df.rename(columns={'count_per_min':'msg_rate_per_min'}, inplace=True)

# ---------- pick rows to be malicious ----------
m_count = int(n * MALICIOUS_FRACTION)
malicious_idx = df.sample(n=m_count, random_state=RANDOM_SEED).index.tolist()
df.loc[malicious_idx, 'is_malicious'] = 1

random.shuffle(malicious_idx)
start = 0
def pick_chunk(frac):
    global start
    k = max(1, int(m_count * frac))
    chunk = malicious_idx[start:start+k]
    start += k
    return chunk

gps_idx = pick_chunk(GPS_SPOOF_FRAC)
replay_idx = pick_chunk(REPLAY_FRAC)
fab_idx = pick_chunk(FABRICATION_FRAC)
sybil_idx = pick_chunk(SYBIL_FRAC)
ts_idx = pick_chunk(TIMESTAMP_SHIFT_FRAC)
spat_idx = pick_chunk(SPAT_SPOOF_FRAC)
dos_idx = pick_chunk(DOS_FRAC)
sub_idx = pick_chunk(SUBTLE_MOD_FRAC)
leftover = malicious_idx[start:]
gps_idx += leftover

# ---------- ATTACK SIMULATORS ----------
# 1) GPS spoofing
for i in gps_idx:
    lat = df.at[i,'coreData_position_lat']
    lon = df.at[i,'coreData_position_long']
    if pd.isna(lat) or pd.isna(lon): 
        continue
    dist = np.random.uniform(SPOOF_DISTANCE_MIN_M, SPOOF_DISTANCE_MAX_M)
    bearing = np.random.uniform(0,360)
    newlat,newlon = move_point(lat, lon, bearing, dist)
    df.at[i, 'coreData_position_lat'] = newlat
    df.at[i, 'coreData_position_long'] = newlon
    df.at[i, 'gps_spoof_flag'] = 1
    df.at[i, 'attack_type'] = 'gps_spoof'

# 2) Replay attack
for i in replay_idx:
    vid = df.at[i, 'coreData_id']
    vehicle_rows = df[(df['coreData_id']==vid) & (df.index < i)]
    if len(vehicle_rows) == 0:
        continue
    chosen = vehicle_rows.sample(n=1, random_state=RANDOM_SEED).iloc[0]
    for col in ['coreData_position_lat','coreData_position_long','coreData_speed','coreData_msgCnt','coreData_secMark']:
        df.at[i, col] = chosen[col]
    df.at[i, 'replay_flag'] = 1
    df.at[i, 'attack_type'] = 'replay'

# 3) Fabrication
for i in fab_idx:
    speed = np.random.uniform(FABRICATED_SPEED_MIN_KMH, FABRICATED_SPEED_MAX_KMH)
    df.at[i, 'coreData_speed'] = speed
    lat = df.at[i,'coreData_position_lat']; lon = df.at[i,'coreData_position_long']
    if not pd.isna(lat):
        br = np.random.uniform(0,360); d = np.random.uniform(500, 5000)
        newlat,newlon = move_point(lat, lon, br, d)
        df.at[i,'coreData_position_lat'] = newlat
        df.at[i,'coreData_position_long'] = newlon
    df.at[i, 'fabrication_flag'] = 1
    df.at[i, 'attack_type'] = 'fabrication'

# 4) Sybil
next_new_id = df['coreData_id'].max() + 1 if df['coreData_id'].dtype.kind in 'iu' else None
sybil_rows_to_append = []
for i in sybil_idx:
    lat = df.at[i,'coreData_position_lat']; lon = df.at[i,'coreData_position_long']
    time = df.at[i,'metadata_generatedAt']
    if pd.isna(lat) or pd.isna(lon): continue
    count = np.random.randint(1,4)
    for k in range(count):
        row = df.loc[i].copy()
        if next_new_id is not None:
            row['coreData_id'] = next_new_id
            next_new_id += 1
        else:
            row['coreData_id'] = f"{row['coreData_id']}_syb_{k}"
        row['is_malicious'] = 1
        row['sybil_flag'] = 1
        row['attack_type'] = 'sybil'
        sybil_rows_to_append.append(row)
if sybil_rows_to_append:
    df = pd.concat([df, pd.DataFrame(sybil_rows_to_append)], ignore_index=True)

# 5) Timestamp shift
for i in ts_idx:
    shift_s = np.random.randint(-300,300)
    df.at[i, 'metadata_generatedAt'] = df.at[i, 'metadata_generatedAt'] + pd.Timedelta(seconds=shift_s)
    df.at[i, 'timestamp_shift_flag'] = 1
    df.at[i, 'attack_type'] = 'timestamp_shift'

# 6) SPAT spoof
if 'payload_data_SPAT_intersections_IntersectionState_status' in df.columns:
    for i in spat_idx:
        cur = df.at[i, 'payload_data_SPAT_intersections_IntersectionState_status']
        if isinstance(cur, str):
            cur_low = cur.lower()
            if 'green' in cur_low:
                new = 'red'
            elif 'red' in cur_low:
                new = 'green'
            elif 'yellow' in cur_low:
                new = 'red'
            else:
                new = 'unknown'
            df.at[i, 'payload_data_SPAT_intersections_IntersectionState_status'] = new
            df.at[i, 'spat_spoof_flag'] = 1
            df.at[i, 'attack_type'] = 'spat_spoof'

# 7) DoS/flood
dos_rows_to_append = []
for i in dos_idx:
    vid = df.at[i,'coreData_id']
    time = df.at[i,'metadata_generatedAt']
    lat = df.at[i,'coreData_position_lat']; lon = df.at[i,'coreData_position_long']
    if pd.isna(lat) or pd.isna(time): continue
    burst = np.random.randint(10,30)
    for b in range(burst):
        newrow = df.loc[i].copy()
        newrow['metadata_generatedAt'] = time + pd.Timedelta(milliseconds=np.random.randint(-500,500))
        newrow['is_malicious'] = 1
        newrow['dos_flag'] = 1
        newrow['attack_type'] = 'dos'
        dos_rows_to_append.append(newrow)
if dos_rows_to_append:
    df = pd.concat([df, pd.DataFrame(dos_rows_to_append)], ignore_index=True)

# 8) Subtle modification
for i in sub_idx:
    if 'coreData_brakes_wheelBrakes_leftFront' in df.columns:
        cur = df.at[i,'coreData_brakes_wheelBrakes_leftFront']
        try:
            df.at[i,'coreData_brakes_wheelBrakes_leftFront'] = 0 if cur==1 else 1
        except:
            pass
    if 'coreData_heading' in df.columns:
        df.at[i,'coreData_heading'] = (df.at[i,'coreData_heading'] + np.random.uniform(90,180)) % 360
    df.at[i,'subtle_mod_flag'] = 1
    df.at[i,'attack_type'] = 'subtle_mod'

# ---------- recompute derived features ----------
df = df.sort_values(['coreData_id','metadata_generatedAt']).reset_index(drop=True)
df['prev_lat'] = df.groupby('coreData_id')['coreData_position_lat'].shift(1)
df['prev_long'] = df.groupby('coreData_id')['coreData_position_long'].shift(1)
df['prev_time'] = df.groupby('coreData_id')['metadata_generatedAt'].shift(1)
df['pos_jump_m'] = df.apply(compute_pos_jump, axis=1)
df['time_delta_s'] = (df['metadata_generatedAt'] - df['prev_time']).dt.total_seconds()
df['computed_speed_kmh'] = (df['pos_jump_m'] / df['time_delta_s']).replace([np.inf, -np.inf], np.nan) * 3.6
df['speed_inconsistent_kmh'] = np.abs(df['coreData_speed'] - df['computed_speed_kmh'])
df['duplicate_msg'] = df.duplicated(subset=['coreData_id','coreData_msgCnt','coreData_secMark'], keep=False).astype(int)
df['msg_epoch_min'] = df['metadata_generatedAt'].dt.floor('T')
msg_rate = df.groupby(['coreData_id','msg_epoch_min']).size().reset_index(name='count_per_min')
df = df.merge(msg_rate, on=['coreData_id','msg_epoch_min'], how='left')
df.rename(columns={'count_per_min':'msg_rate_per_min'}, inplace=True)
df['loc_round_lat'] = df['coreData_position_lat'].round(5)
df['loc_round_long'] = df['coreData_position_long'].round(5)
dup_ids = df.groupby(['loc_round_lat','loc_round_long','msg_epoch_min'])['coreData_id'].nunique().reset_index(name='unique_ids_here')
df = df.merge(dup_ids, on=['loc_round_lat','loc_round_long','msg_epoch_min'], how='left')
df.rename(columns={'unique_ids_here':'duplicate_id_count'}, inplace=True)

# ---------- unify and clean attack columns ----------
df.loc[(df['is_malicious'] == 1) & (df['attack_type'].isna()), 'attack_type'] = 'unknown'
df.loc[df['is_malicious'] == 0, 'attack_type'] = 'none'

attack_flag_cols = [
    'gps_spoof_flag','replay_flag','fabrication_flag','sybil_flag',
    'timestamp_shift_flag','spat_spoof_flag','dos_flag','subtle_mod_flag'
]
df = df.drop(columns=[c for c in attack_flag_cols if c in df.columns])

# ---------- housekeeping ----------
indicator_cols = [
    'is_malicious','attack_type','pos_jump_m','time_delta_s','computed_speed_kmh',
    'speed_inconsistent_kmh','duplicate_msg','msg_rate_per_min','duplicate_id_count'
]
present_cols = [c for c in indicator_cols if c in df.columns]
print("Indicator columns present:", present_cols)

print(df[present_cols].sample(10, random_state=RANDOM_SEED))

merged_df_with_attacks = df.copy()
merged_df_with_attacks.to_csv('merged_with_attacks.csv', index=False)
print("✅ Saved merged_with_attacks.csv (with unified attack column)")
```

```python
# ===================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
import shap
import gc
import warnings
import sys

# -----------------------------------
# Step 0: Suppress LightGBM warnings and logs
# -----------------------------------
warnings.filterwarnings("ignore")

# Define a dummy writer to silence stdout temporarily
class NullWriter:
    def write(self, arg): pass
    def flush(self): pass

null_writer = NullWriter()

# -----------------------------------
# Step 1: Load dataset
# -----------------------------------
df = pd.read_csv("final_merged_dataset_with_attacks.csv")
print("Loaded sample shape:", df.shape)

target_col = "attack_type"

# Drop duplicates and missing values
df = df.drop_duplicates().dropna(subset=[target_col])

# Keep only valid, unique records (but don’t discard duplicate info completely)
df = df.drop_duplicates(subset=["coreData_id", "coreData_secMark"], keep="first")

# -----------------------------------
# Step 2: Hybrid reliability feature engineering
# -----------------------------------
# Handle missing duplicate fields gracefully
if "duplicate_msg" not in df.columns:
    df["duplicate_msg"] = 0
if "duplicate_id_count" not in df.columns:
    df["duplicate_id_count"] = 1

# Create hybrid reliability metrics
df["duplicate_msg_flag"] = df["duplicate_msg"].astype(int)
df["duplicate_id_ratio"] = df["duplicate_id_count"] / (
    df.groupby("coreData_id")["duplicate_id_count"].transform("max") + 1e-6
)

# Optional interaction feature: reliability-weighted speed (if speed exists)
if "coreData_speed" in df.columns:
    df["speed_weighted_by_reliability"] = df["coreData_speed"] * (1 - df["duplicate_id_ratio"])

# -----------------------------------
# Step 3: Separate features and target
# -----------------------------------
y = df[target_col]
features_df = df.drop(columns=[target_col, 'is_malicious'], errors='ignore')

# -----------------------------------
# Step 4: Identify column types
# -----------------------------------
categorical_columns = features_df.select_dtypes(include=["object", "category"]).columns.tolist()
numeric_columns = features_df.select_dtypes(include=[np.number]).columns.tolist()

# -----------------------------------
# Step 5: Encode target numerically
# -----------------------------------
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# -----------------------------------
# Step 6: Target encoding for categorical columns
# -----------------------------------
top_cats = sorted(categorical_columns, key=lambda c: features_df[c].nunique())[:10]

temp_df = features_df.copy()
temp_df[target_col] = y_encoded

for col in top_cats:
    encoding = temp_df.groupby(col)[target_col].mean()
    features_df[col] = features_df[col].map(encoding)
    features_df[col].fillna(features_df[col].mean(), inplace=True)

# -----------------------------------
# Step 7: Combine hybrid numeric + target-encoded features
# -----------------------------------
hybrid_features = numeric_columns + top_cats
features_df = features_df[hybrid_features].astype(np.float32)

print(f"Training samples: {features_df.shape[0]}, features: {features_df.shape[1]}")

# -----------------------------------
# Step 8: Train/test split
# -----------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    features_df, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# -----------------------------------
# Step 9A: Train LightGBM model (quiet mode)
# -----------------------------------
print("\nTraining LightGBM model...")

lgb_model = LGBMClassifier(
    n_estimators=200,
    max_depth=8,
    learning_rate=0.05,
    num_leaves=64,
    subsample=0.8,
    colsample_bytree=0.8,
    n_jobs=-1,
    verbose=-1
)

# Temporarily suppress stdout
old_stdout = sys.stdout
sys.stdout = null_writer
lgb_model.fit(X_train, y_train)
sys.stdout = old_stdout

# -----------------------------------
# Step 9B: Train XGBoost model
# -----------------------------------
print("\nTraining XGBoost model...")

xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method='hist',
    eval_metric='mlogloss',
    n_jobs=-1,
    verbosity=0
)
xgb_model.fit(X_train, y_train)

# -----------------------------------
# Step 10: Evaluation
# -----------------------------------
print("\nEvaluating models...")

y_pred_lgb = lgb_model.predict(X_test)
y_pred_xgb = xgb_model.predict(X_test)

print("\n=== LightGBM Classification Report ===")
print(classification_report(y_test, y_pred_lgb, target_names=label_encoder.classes_))

print("\n=== XGBoost Classification Report ===")
print(classification_report(y_test, y_pred_xgb, target_names=label_encoder.classes_))

# -----------------------------------
# Step 11: SHAP explanation for XGBoost (sampled)
# -----------------------------------
print("\nComputing SHAP values for XGBoost (sampled)...")

X_background = X_train.sample(min(300, len(X_train)), random_state=42)
X_eval = X_test.sample(min(300, len(X_test)), random_state=42)

explainer = shap.Explainer(xgb_model, X_background)
shap_values = explainer(X_eval, check_additivity=False)
shap.summary_plot(shap_values, X_eval, show=False)

# -----------------------------------
# Step 12: Memory cleanup
# -----------------------------------
del temp_df, shap_values, X_train, X_test, y_train, y_test
gc.collect()
```

### Block chain security

```python
import hashlib
import json
import pandas as pd
from datetime import datetime
from typing import List


# ==========================
# 1. Blockchain Classes
# ==========================
class Block:
    def __init__(self, index, timestamp, data, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.compute_hash()

    def compute_hash(self):
        block_string = json.dumps({
            "index": self.index,
            "timestamp": self.timestamp,
            "data": self.data,
            "previous_hash": self.previous_hash
        }, sort_keys=True).encode()
        return hashlib.sha256(block_string).hexdigest()


class Blockchain:
    def __init__(self):
        self.chain: List[Block] = []
        self.create_genesis_block()

    def create_genesis_block(self):
        genesis_block = Block(0, str(datetime.utcnow()), {"msg": "Genesis Block"}, "0")
        self.chain.append(genesis_block)

    def get_last_block(self):
        return self.chain[-1]

    def add_block(self, data):
        prev_block = self.get_last_block()
        new_block = Block(len(self.chain), str(datetime.utcnow()), data, prev_block.hash)
        self.chain.append(new_block)
        return new_block


# ==========================
# 2. RSU Node Definition
# ==========================
class RSUNode:
    def __init__(self, node_id, blockchain):
        self.node_id = node_id
        self.blockchain = blockchain

    def preprocess_data(self, df: pd.DataFrame):
        """Normalize coordinates and speed values"""
        # Avoid overwriting original data by working on a copy
        df = df.copy()

        # Normalize latitude/longitude if too large
        if df['coreData_position_lat'].abs().max() > 1000:
            df['coreData_position_lat'] /= 1e7
            df['coreData_position_long'] /= 1e7

        # Normalize speed if it looks too high (raw decimeters/sec)
        if df['coreData_speed'].mean() > 200:
            df['coreData_speed'] = df['coreData_speed'] / 100.0

        return df

    def validate_and_commit(self, df: pd.DataFrame):
        """Preprocess, validate, and add block"""
        df = self.preprocess_data(df)
        validated = self.validate_messages(df)
        trust_score = self.compute_trust_metrics(validated)

        block_data = {
            "node_id": self.node_id,
            "validated_count": len(validated),
            "trust_score": trust_score,
            "messageRoot": hashlib.sha256(validated.to_json().encode()).hexdigest(),
            "timestamp": str(datetime.utcnow())
        }
        return self.blockchain.add_block(block_data)

    def validate_messages(self, df: pd.DataFrame):
        """Drop duplicates and invalid coordinate/speed values"""
        df = df.drop_duplicates(subset=['coreData_id', 'coreData_msgCnt'])
        df = df[df['coreData_speed'] >= 0]
        df = df[df['coreData_position_lat'].between(-90, 90)]
        df = df[df['coreData_position_long'].between(-180, 180)]
        return df

    def compute_trust_metrics(self, df: pd.DataFrame):
        """Compute network trust indicators"""
        malicious_rate = df['is_malicious'].mean() if 'is_malicious' in df else 0
        avg_speed = df['coreData_speed'].mean() if 'coreData_speed' in df else 0
        anomaly_rate = df['duplicate_id_count'].mean() if 'duplicate_id_count' in df else 0

        return {
            "avg_speed_kmh": round(avg_speed, 2),
            "malicious_rate": round(malicious_rate, 3),
            "anomaly_rate": round(anomaly_rate, 3)
        }


# ==========================
# 3. Load Your Dataset
# ==========================
DATA_PATH = "final_merged_dataset_with_attacks.csv"  # <-- Replace with your actual CSV path

columns_to_use = [
    "coreData_id",
    "coreData_msgCnt",
    "coreData_speed",
    "coreData_position_lat",
    "coreData_position_long",
    "computed_speed_kmh",
    "is_malicious",
    "duplicate_id_count"
]

df = pd.read_csv(DATA_PATH, usecols=columns_to_use, low_memory=False)
print(f"\n✅ Loaded dataset: {df.shape[0]} records, {df.shape[1]} columns")

# ==========================
# 4. Run RSU Node + Blockchain
# ==========================
blockchain = Blockchain()
rsu = RSUNode("RSU-Abuja", blockchain)

validated_block = rsu.validate_and_commit(df)

# ==========================
# 5. Display Blockchain Summary
# ==========================
print("\n=== Blockchain Ledger Summary ===")
for block in blockchain.chain:
    print(json.dumps(block.__dict__, indent=2))

print("\n✅ Blockchain simulation completed.")
```
